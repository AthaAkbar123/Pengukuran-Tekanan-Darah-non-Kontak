```python
import json
import numpy as np
import os

def process_json_file(json_file):
    with open(json_file, 'r') as file:
        data = json.load(file)
    
    # Ekstrak semua nilai o2saturation
    o2_saturation_values = [entry["Value"]["o2saturation"] for entry in data["/FullPackage"]]
    
    return o2_saturation_values

# Ambil semua folder dalam direktori saat ini
base_folder = os.path.join(os.getcwd(), "PURE Dataset")
folders = [folder for folder in os.listdir(base_folder) if os.path.isdir(os.path.join(base_folder, folder))]

folder_o2_data = {}
folder_means = {}

for folder in folders:
    folder_path = os.path.join(base_folder, folder)
    json_file = os.path.join(folder_path, f"{folder}.json")
    if os.path.exists(json_file):
        values = process_json_file(json_file)
        folder_o2_data[folder] = values
        folder_means[folder] = np.mean(values) if values else 0

# Hitung rata-rata per folder
o2_saturation_all = [value for values in folder_o2_data.values() for value in values]

ground_truth_per_folder = {folder: np.mean(values) for folder, values in folder_o2_data.items() if values}

# Tampilkan hasil rata-rata per folder
print("Ground Truth O₂ Saturation per Folder:")
for folder, mean_value in ground_truth_per_folder.items():
    print(f"{folder}: {mean_value:.2f}%")

# Hitung keseluruhan statistik
ground_truth_o2 = np.mean(o2_saturation_all)
std_o2 = np.std(o2_saturation_all)

print(f"\nOverall Ground Truth O₂ Saturation: {ground_truth_o2:.2f}%")
print(f"Standar Deviasi: {std_o2:.2f}%")
print(f"Rentang Nilai: {min(o2_saturation_all)}% - {max(o2_saturation_all)}%")
```

```python
import json
import numpy as np
import os
import pandas as pd

def process_json_file(json_file):
    with open(json_file, 'r') as file:
        data = json.load(file)
    
    # Ekstrak semua nilai o2saturation
    o2_saturation_values = [entry["Value"]["o2saturation"] for entry in data["/FullPackage"]]
    
    return o2_saturation_values

# Ambil semua folder dalam direktori saat ini
base_folder = os.path.join(os.getcwd(), "PURE Dataset")
folders = [folder for folder in os.listdir(base_folder) if os.path.isdir(os.path.join(base_folder, folder))]

folder_o2_data = {}
folder_means = {}

for folder in folders:
    folder_path = os.path.join(base_folder, folder)
    json_file = os.path.join(folder_path, f"{folder}.json")
    if os.path.exists(json_file):
        values = process_json_file(json_file)
        folder_o2_data[folder] = values
        folder_means[folder] = np.mean(values) if values else 0

# Hitung rata-rata per folder
o2_saturation_all = [value for values in folder_o2_data.values() for value in values]

ground_truth_per_folder = {folder: np.mean(values) for folder, values in folder_o2_data.items() if values}

# Buat DataFrame untuk hasil folder
folder_data = {
    'Folder': list(ground_truth_per_folder.keys()),
    'Rata-rata O₂ Saturation (%)': list(ground_truth_per_folder.values())
}

# Buat DataFrame untuk statistik keseluruhan
overall_stats = {
    'Statistik': ['Overall Rata-rata O₂ Saturation', 'Standar Deviasi', 'Rentang Nilai'],
    'Nilai': [
        np.mean(o2_saturation_all),
        np.std(o2_saturation_all),
        f"{min(o2_saturation_all)}% - {max(o2_saturation_all)}%"
    ]
}

# Membuat DataFrame untuk folder dan keseluruhan statistik
folder_df = pd.DataFrame(folder_data)
overall_df = pd.DataFrame(overall_stats)

# Tentukan file output Excel
output_file = "o2_saturation_statistics.xlsx"

# Simpan ke dalam file Excel dengan dua sheet
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    folder_df.to_excel(writer, sheet_name='Folder Statistik', index=False)
    overall_df.to_excel(writer, sheet_name='Keseluruhan Statistik', index=False)

print(f"Results have been saved to {output_file}")
```
